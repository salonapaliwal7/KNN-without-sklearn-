```python
# Separate features and target

X = df.drop('class', axis=1)
y= df['class']
```

```python
# train-test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=100)
```

```python
# Scaling

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
y_train = y_train.values
y_test = y_test.values
```

```python
# euclidian dist to do KNN

def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))
```

```python
# KNN without sklearn

def KNN(X_train, y_train, x_test, k=5):
  distances = []
  for i in range(len(X_train)):
    dist = euclidean_distance(x_test,X_train[i])
    distances.append((dist, y_train[i])) # append dist and its corresponding class

  # sort distances
  distances.sort(key = lambda x:x[0]) # sort based on first value in tuple

  # pick top-K
  k_neighbors = distances[:k]

  # extract labels
  labels = [label for _, label in k_neighbors]

  # majority vote
  return max(set(labels), key=labels.count)

```

```python
# Predict for all data points

def predict(X_train, y_train, X_test, k=5):
    predictions = []

    for x in X_test:
        pred = KNN(X_train, y_train, x, k) # call KNN for each data point
        predictions.append(pred)
    return np.array(predictions)
```

```python
# find Accuracy

def accuracy(y_true, y_pred):
    return np.sum(y_true == y_pred) / len(y_true)
```

```python
y_pred = predict(X_train, y_train, X_test, k=5)
print("Accuracy:", accuracy(y_test, y_pred))
```

# WEIGHTED KNN AFTER DOING PCA


```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df = pd.read_csv("binary_class.csv")

X= df.drop('class', axis=1)
y= df['class']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state=100)

X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

# fills the missing columns values with 0, so there is consistency in train & test data
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)

pca = PCA(n_components=3)
X_train = pca.fit_transform(X_train)
X_test = pca.fit_transform(X_test)
```

```python
def euclidean_dist(x1,x2):
  return np.sqrt(np.sum(x1-x2)**2)


# Weighted KNN
def weighted_knn(X_train, y_train, X_test, k=5):
  distances = []
  for i in range(len(X_train)):
    dist = euclidean_dist(X_test,X_train[i])
    distances.append((dist, y_train[i]))

  distances.sort(key=lambda x:x[0])

  k_neighbors = distances[:k]

  weights = {}

  for dist, label in k_neighbors:
    weight = (1/(dist+1e-8))

    if label in  weights:
      weights[label] +=1

    else:
      weights[label] = 1

  return max(weights, key=weights.get)
```

```python
def predict_weighted(X_train, y_train, X_test, k=5):
    predictions = []

    for x in X_test:
        pred = weighted_knn(X_train, y_train, x, k)
        predictions.append(pred)

    return np.array(predictions)


def accuracy(y_true, y_pred):
    return np.sum(y_true == y_pred) / len(y_true)



y_pred = predict_weighted(X_train, y_train.values, X_test, k=5)

print("Accuracy:", accuracy(y_test.values, y_pred))
```

- Weighted KNN assumes:

  👉 closer points = more reliable

- Weighted KNN gives:

  👉 very close point → VERY high weight




🔥 Intuition

Instead of:

👉 “majority wins”

You now say:

👉 “closer neighbors matter more”



“Weighted KNN improves standard KNN by assigning higher influence to closer neighbors, reducing the impact of noisy distant points.”
